# Parte 2. Entornos complejos

Este notebook constituye la **puerta de entrada** a la parte del trabajo dedicada al estudio de métodos de **aprendizaje por refuerzo en entornos más complejos**.  
Aquí se organiza la ejecución y la lectura de los experimentos realizados con distintos enfoques, incluyendo métodos de **Monte Carlo**, **métodos tabulares** y **métodos aproximados de control**.

El objetivo general de esta parte es analizar el comportamiento de estos métodos en escenarios de mayor complejidad, observando sus características, su desempeño y las diferencias entre enfoques tabulares y aproximados.


## Organización de esta parte

Los notebooks incluidos en esta sección son los siguientes:

- **`A_MonteCarloTodasLasVisitas.ipynb`**: desarrollo y experimentación con el método Monte Carlo de todas las visitas.
- **`B_metodos_tabulares.ipynb`**: estudio e implementación de métodos tabulares de control.
- **`B_control_aprox.ipynb`**: desarrollo y análisis de métodos aproximados de control.

> **Nota:** esta organización permite recorrer la parte de manera progresiva, pasando desde enfoques más clásicos hacia métodos de mayor complejidad.
