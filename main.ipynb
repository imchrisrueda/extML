## Preparación del entorno

La siguiente celda permite identificar el entorno de ejecución y verificar que la estructura principal del repositorio esté disponible.  
No ejecuta experimentos, pero ayuda a confirmar que las carpetas esperadas pueden localizarse correctamente tanto en ejecución local como en Google Colab.


```python
# Preparación básica del entorno para ejecución local o en Google Colab.

import sys
from pathlib import Path

IN_COLAB = "google.colab" in sys.modules

if IN_COLAB:
    print("Entorno detectado: Google Colab")
else:
    print("Entorno detectado: ejecución local")

repo_root = Path.cwd()
print(f"Directorio actual: {repo_root}")

if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))
    print("Se agregó la raíz del repositorio a sys.path")
else:
    print("La raíz del repositorio ya estaba en sys.path")
```
