# Parte 1. Bandido de \(k\)-brazos

Este notebook constituye la **puerta de entrada** a la parte del trabajo dedicada al problema del **bandido de \(k\)-brazos**.  
Aquí se organiza la ejecución y la lectura de los experimentos realizados con distintas estrategias de selección de acciones, estudiando su comportamiento sobre distribuciones de recompensa de tipo **Normal**, **Bernoulli** y **Binomial**.

El objetivo general de esta parte es comparar el desempeño de los métodos en términos de:

- **recompensa promedio**,
- **porcentaje de selección del brazo óptimo**,
- **regret acumulado**.

Este notebook no desarrolla los experimentos en detalle, sino que orienta la navegación del repositorio y deja preparado el entorno para ejecutar los notebooks específicos de cada método.


## Organización de esta parte

Los notebooks incluidos en esta sección son los siguientes:

- **`banditGreedy.ipynb`**: desarrollo y experimentación con la estrategia \(\varepsilon\)-greedy.
- **`banditSoftmax.ipynb`**: desarrollo y experimentación con la estrategia Softmax.
- **`banditUCB.ipynb`**: desarrollo y experimentación con las variantes UCB implementadas.
- **`banditComparacion.ipynb`**: comparación final entre las mejores configuraciones obtenidas para cada estrategia.

> **Nota:** este `main.ipynb` asume que el notebook de comparación final fue renombrado como `banditComparacion.ipynb`.
